```python
from evds import evdsAPI
import matplotlib.pyplot as plt

evds = evdsAPI('Er16YO2Bhy')

#takes all the data in one dataframe
data=evds.get_data(['TP.YISGUCU2.G8','TP.DK.USD.S.YTL','TP.FG.J0'], startdate="01-01-2018", enddate="01-01-2021", frequency=5)

#takes data by percentage change
dataperc=evds.get_data(['TP.YISGUCU2.G8','TP.DK.USD.S.YTL','TP.FG.J0'], startdate="01-01-2018", enddate="01-01-2021", frequency=5, formulas=1)

#takes all three data seperately

#1 Unemployement rate
data1=evds.get_data(['TP.YISGUCU2.G8'], startdate="01-01-2018", enddate="01-01-2021", frequency=5)

#2 USD/TRY 
data2=evds.get_data(['TP.DK.USD.S.YTL'], startdate="01-01-2018", enddate="01-01-2021", frequency=5)

#3 Consumer Price Index
data3=evds.get_data(['TP.FG.J0'], startdate="01-01-2018", enddate="01-01-2021", frequency=5)

data
```

```python

#plots 3 data seperately
data1.plot()
plt.xlabel('Months')
plt.legend(loc='upper left')
plt.title("Unemployement Rate in Turkey")

data2.plot()
plt.xlabel('Months')
plt.legend(loc='upper left')
plt.title("USD/TRY")

data3.plot()
plt.xlabel('Months')
plt.legend(loc='upper left')
plt.title("Consumer Price Index of Turkey")

#plots all data's ny percentage change
dataperc.plot()
plt.legend(bbox_to_anchor =(1.20, 1.35), ncol = 3)
plt.xlabel('Months')
plt.title('Percentage changes of all data above')

```

```python
"""

As we see, there is an increasing trend about USD/TRY and Consumer Price Index. Prices effected by exchange rates 
directly proportional. Also, we can see the similarity between exchange rate and unemployement rate but not similar as CPI.
We can say that Dollar don't effect unemployement rate simultanously.

"""
```
